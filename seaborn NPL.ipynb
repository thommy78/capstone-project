```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("npl_tidy.csv")
df1=pd.read_csv("npl_melted.csv")
```

```python
sns.set_style("ticks")
sns.relplot(x="Date", y='NPL',
data=df1,
kind="line", hue='Country', height=5, aspect=3)
plt.title('NPL Development')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=df, whis=[10, 90]).set_title('NPL distribution')
plt.show()
```
